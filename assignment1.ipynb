3.Compare and contrast mutable and immutable objects in python with examples.

```python
## Mutable vs Immutable Objects in Python

###  Mutability

#Mutable objects can be changed after they are created. Their internal state can be modified without creating a new object.
 #Examples: lists, dictionaries, sets.

### Immutability

#Immutable objects cannot be changed once created. Any modification creates a new object.
# Examples: numbers (integers, floats), strings, tuples, booleans.

### Comparison Table

# | Feature       | Mutable Objects                                                  | Immutable Objects |
# | Changeability | Can be changed after creation                                    | Cannot be changed after creation |
# | Examples      | Lists, dictionaries, sets                                        | Numbers, strings, tuples, booleans |
# | Use Cases     | Dynamic data structures, where frequent modifications are needed | When data integrity is crucial, or when you need to ensure consistency |

### Examples

#### Mutable Objects

# List
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

# Modify the list
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

# Dictionary
my_dict = {'a': 1, 'b': 2}
print(my_dict)  # Output: {'a': 1, 'b': 2}

# Modify the dictionary
my_dict['c'] = 3
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}


#### Immutable Objects

#python
# String
my_string = "hello"
print(my_string)  # Output: hello

# Trying to modify a string will raise an error
# my_string[0] = 'H'  # TypeError: 'str' object does not support item assignment

# Integer
num = 10
print(num)  # Output: 10

# Trying to modify an integer will also raise an error
# num += 1  # This creates a new integer object with the value 11

### Key Points

# Performance: Immutable objects are often faster for access but can be less efficient for frequent modifications.
# Data Integrity: Immutable objects provide data integrity as they cannot be accidentally changed.
# Memory Usage: Immutable objects can sometimes lead to increased memory usage due to the creation of new objects.
# Hashing: Immutable objects are hashable, making them suitable for use as dictionary keys.
```

4.Discuss the different types of operators in python and provide examples of how they are used.

```python
## Python Operators

#Operators are special symbols that perform specific operations on one or more values (operands). Python supports a variety of operators for different purposes. Let's explore them with examples:

### Arithmetic Operators
# These operators perform basic mathematical operations.
# +: Addition (e.g., `x + y`)
# -: Subtraction (e.g., `x - y`)
#  : Multiplication (e.g., `x   y`)
# /: Division (e.g., `x / y`)
# %: Modulus (remainder) (e.g., `x % y`)
# //: Floor division (quotient without remainder) (e.g., `x // y`)
#   : Exponentiation (e.g., `x    y`)

x = 10
y = 3
print(x + y)  # Output: 13
print(x - y)  # Output: 7
print(x * y)  # Output: 30
print(x / y)  # Output: 3.3333333333333335
print(x % y)  # Output: 1
print(x // y)  # Output: 3
print(x **  y)  # Output: 1000

### Comparison Operators
# These operators compare two values and return a Boolean result (True or False).
# ==: Equal to
# !=: Not equal to
# >: Greater than
# <: Less than
# >=: Greater than or equal to
# <=: Less than or equal to

x = 10
y = 5
print(x == y)  # Output: False
print(x != y)  # Output: True
print(x > y)  # Output: True
print(x < y)  # Output: False
print(x >= y)  # Output: True
print(x <= y)  # Output: False

### Logical Operators
# These operators combine conditional statements.
# and: Both conditions must be true
# or: At least one condition must be true
# not: Inverts the truth value

x = True
y = False
print(x and y)  # Output: False
print(x or y)  # Output: True
print(not x)  # Output: False

### Assignment Operators
# These operators assign values to variables.
# =: Simple assignment
# +=: Add and assign
# -=: Subtract and assign
# *=: Multiply and assign
# /=: Divide and assign
# %=: Modulus and assign
# //=: Floor division and assign
# **=: Exponentiation and assign

x = 5
x += 3  # Equivalent to x = x + 3
print(x)  # Output: 8


### Bitwise Operators
# These operators perform bit-level operations on integers.
# &: Bitwise AND
# |: Bitwise OR
# ^: Bitwise XOR
# ~: Bitwise NOT
# <<: Left shift
# >>: Right shift

x = 5  # Binary: 0101
y = 3  # Binary: 0011
print(x & y)  # Output: 1 (Binary: 0001)


### Membership Operators
# These operators test for membership in a sequence.
# in: Checks if a value is present in a sequence
# not in: Checks if a value is not present in a sequence


my_list = [1, 2, 3]
print(2 in my_list)  # Output: True
print(4 not in my_list)  # Output: True


### Identity Operators
# These operators compare the identity of two objects.
#     is  : Checks if two objects are the same object
#     is not  : Checks if two objects are not the same object


x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)  # Output: True
print(x is z)  # Output: False


#These are the primary types of operators in Python. Understanding them is crucial for effective programming.
```

5.Explain the concept of typecasting in python with examples.

```python
#Type casting is the process of converting one data type into another. Python supports both implicit and explicit type casting.

### Implicit Type Casting
#Python automatically converts one data type to another when it is necessary to prevent data loss. This is known as implicit type casting or coercion.

x = 10  # Integer
y = 3.14  # Float

# Python automatically converts x to a float for the division operation
result = x / y
print(result)  # Output: 3.184713375796178

#In this example, `x` is an integer, but it's automatically converted to a float before the division operation because dividing an integer by a float results in a float.

### Explicit Type Casting
#You can manually convert a data type to another using built-in functions like `int()`, `float()`, and `str()`. This is known as explicit type casting.

#Example

x = 10.5
y = "30"

# Convert float to integer
a = int(x)
print(a)  # Output: 10

# Convert string to integer
b = int(y)
print(b)  # Output: 30

# Convert integer to string
c = str(a)
print(c)  # Output: "10"

# **Important Notes:**

# * When converting from a float to an integer, the decimal part is truncated.
# * Converting a string to an integer or float is only possible if the string represents a valid number. Otherwise, it will raise a `ValueError`.
# * Converting a number to a string is always possible.

# **Common Type Casting Functions:**

# int(): Converts a value to an integer.
# float(): Converts a value to a float.
# str(): Converts a value to a string.
# bool(): Converts a value to a boolean (True or False).
# list(): Converts an iterable to a list.
# tuple(): Converts an iterable to a tuple.
# set(): Converts an iterable to a set.
# dict(): Converts a sequence of key-value pairs to a dictionary.

# By understanding type casting, we can manipulate data types effectively in your Python programs.

```

6.How do conditional statement works in python? illustrate with examples.

```python
#Conditional statements allow you to control the flow of your program based on certain conditions. Python provides three main conditional statements: `if`, `else`, and `elif`.

### The `if` Statement
#The `if` statement executes a block of code if a specified condition is true.

number = 10

if number > 0:
  print("The number is positive.")

### The `if-else` Statement
#The `if-else` statement executes one block of code if a condition is true, and another block of code if the condition is false.

number = -5

if number > 0:
  print("The number is positive.")
else:
  print("The number is negative or zero.")

### The `elif` Statement
#The `elif` statement allows you to check multiple conditions sequentially.

number = 0

if number > 0:
  print("The number is positive.")
elif number == 0:
  print("The number is zero.")
else:
  print("The number is negative.")

### Nested `if` Statements
#You can nest `if` statements within other `if` statements to create more complex conditions.

x = 42
y = 20

if x > y:
  print("x is greater than y")
  if x > 30:
    print("x is also greater than 30")
else:
  print("x is less than or equal to y")

### Conditional Expressions
#Python also supports conditional expressions, which are a shorthand way of writing `if-else` statements in a single line.

x = 10
y = 5
result = x if x > y else y
print(result)  # Output: 10

# Key points to remember:

#  Indentation is crucial in Python to define code blocks within conditional statements.
#  we can use comparison operators (e.g., `>`, `<`, `==`, `!=`, `>=`, `<=`) and logical operators (`and`, `or`, `not`) to create conditions.
#  The `elif` statement is an abbreviation of "else if".
#  Nested `if` statements can be used for complex decision-making, but they should be used carefully to avoid making code hard to read.

#By understanding and effectively using conditional statements, we can create Python programs that make decisions and perform different actions based on various conditions.

```

7.Describe the different types of loops in python and their use cases in python.

```python
#Loops are fundamental constructs in programming that allow you to execute a block of code repeatedly. Python offers two primary types of loops: `for` and `while`.

### For Loop

# A `for` loop is used to iterate over a sequence (like a list, tuple, string, or dictionary) or any other iterable object.

#for item in iterable:
  # code to be executed

### While Loop

# A `while` loop repeatedly executes a block of code as long as a given condition is true.

#while condition:
  # code to be executed

#Example:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit) # Output: apple banana cherry

# Use cases:
# Iterating over lists, tuples, strings
# Processing elements in a sequence
# Creating new lists or data structures based on existing ones

### While Loop
#A `while` loop repeatedly executes a block of code as long as a given condition is true.

#while condition:
  # code to be executed

#Example:
count = 0
while count < 5:
  print(count)
  count += 1 # Output: 0 1 2 3 4

# Use cases:
# Repeating actions until a certain condition is met
# Creating menus or interactive programs
# Implementing algorithms that require indefinite iteration

### Nested Loops
# You can create loops within loops, known as nested loops.

#Example:
for i in range(3):
  for j in range(2):
    print(i, j)
    # Output: 0 0
  #          0 1
  #          1 0
  #          1 1
  #          2 0
  #          2 1

#Use cases:
# Processing multi-dimensional data structures
# Creating complex patterns or matrices

### Loop Control Statements
#break: Terminates the loop entirely.
#continue:Skips the current iteration and moves to the next.

#Example:
for num in range(10):
  if num == 5:
    break
  print(num) #Output: 0 1 2 3 4

# Use cases:
# Exiting a loop prematurely based on a condition
# Skipping certain iterations within a loop

### Additional Considerations
# List comprehensions: A concise way to create lists using `for` loops.
# Generator expressions: Similar to list comprehensions, but create generators instead of lists.

#By understanding these loop types and control statements, we can effectively handle repetitive tasks and create efficient Python programs.

```
